10. Use the zscore function to standardize a numerical field (create a new column), then show its
histogram

```python
from scipy import stats
```

```python
election["availableMandates_z"] = stats.zscore(election['availableMandates'])
```

```python
print(election['availableMandates_z'].head(n=20))
```

```python
election['availableMandates_z'].plot(kind='hist', title = 'Histogram of available mandates')
```
